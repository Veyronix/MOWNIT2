### Podstawowe info
#### 1. Regresja w wielu wymiarach:
Bardzo często chcemy dopasować prostą do danych o większej wymiarowości. Przykładowo, cena domu może zależeć od jego powierzchni, ilości pokoi i ilości pięter (w praktyce takich cech danych może być bardzo dużo). Na szczęście idea jest w zasadzie identyczna, jak przy jednowymiarowej regresji z poprzednich zajęć. Zwróćmy uwagę, że gdy umiemy dopasowywać funkcję liniową w wielu wymiarach, możemy robić różne zagraniczne tricki: nikt nie broni nam sztucznie stworzyć sobie nowych zmiennych, będących nieliniowymi przekształceniami istniejących zmiennych. W ten sposób za pomocą jednego algorytmu możemy dopasowywać zarówno funkcje liniowe, jak i np. wielomiany. (Więcej: na zajęciach). <img src="images/dataset.jpg">

Swoją drogą, w internecie wymiennie można znaleźć pojęcia "multi(ple) regression" i "multivariate regression". Można założyć, że pierwsza z nich odnosi się do funkcji z wielowymiarową dziedziną, ale jednowymiarową przeciwdziedziną, natomiast to drugie: z wielowymiarową dziedziną i przeciwdziedziną.

#### 2. **Gradient descent**:
O ile dla małych/prostych zbiorów danych jesteśmy w stanie wyliczyć najlepsze dopasowanie analitycznie (por. MOwNiT 1, laboratorium), o tyle w praktyce zajęłoby to zbyt dużo czasu. Na szczęście minimalizować funkcję błędu możemy heurystycznie, za pomocą poruszania się w kierunku największego spadku. Intuicja: to tak, jakbyśmy chcieli znaleźć najniższe miejsce w jakiejś kotlinie poprzez rzucenie na ziemię metalowej kulki i sprawdzenie, gdzie się zatrzyma.
<img src="images/gradient_descent.png">
Tak prosta metoda może dość łatwo "utknąć" w minimum lokalnym, ale na szczęście funkcje, z którymi mamy do czynienia przy regresji liniowej są bardzo porządne i mają tylko jedno minimum.

#### 3. **Regularyzacja**:
Intuicja podpowiada "im więcej cech danych, tym lepiej je zrozumiemy". Niestety, życie (i konkursy na Kaggle) uczy, że to błędna intuicja. Część zmiennych nie wnosi żadnej informacji, dane są zaszumione, a czasem informacja jest redundantna. Co się dzieje, jeśli dwie cechy są mocno skorelowane? Jeśli są również mocno skorelowane z wartością, którą przewidujemy, zaczniemy tak naprawdę liczyć tą samą cechę, ale podwójnie (to intuicja, nie poprawne tłumaczenie).

### 1. Ładowanie danych
Zaladowanie danych do DataFrame'a

```python
data = pd.read_csv('all/train.csv', index_col=0)
data.head()
```

### 2. Wstępne opisanie danych i brakujące wartości

Jak można zauważyć jest dużo brakujących wartości.

#### Wybranie interesujących danych
Zostało wybrane: rok budowy, wielkość powierzchni apartamentu w stopach^2,
ilość sypialni i cena.
Jak można zauważyc wybrane dane nie posiadają brakujących danych.
Jeśli znalazłyby się dalej jakieś brakujące dane to trzeba byłoby je zastąpić prawdziwymi wartościami,

```python
chosen_data = data[['YearBuilt','LotArea','BedroomAbvGr','SalePrice']]
HTML(chosen_data.describe(include='all').to_html())
```

### 3. Rysowanie wykresów
#### Wybieramy mieszkanie wybudowane po 1950, bo raczej starsze budynki nie są teraz sprzedawane, czyli usuwamy outlinerów.

```python
chosen_data = chosen_data[chosen_data.YearBuilt > 1950]
chosen_data['YearBuilt'].plot.hist(bins=50, alpha=0.5)
```

#### Wybieramy mieszkania z powieżnią mniejszą niz 30000, bo większe mieszkania są outlinerami.

```python
chosen_data = chosen_data[chosen_data.LotArea < 30000]
```

#### Logarytmujemy, aby dane bardziej przypominały rozkład normalny.

```python
chosen_data['LotArea'] = np.log1p(chosen_data["LotArea"])
chosen_data['LotArea'].plot.hist(bins=50, alpha=0.5)
```

#### Tutaj nie ma outlinerów

```python
chosen_data['BedroomAbvGr'].plot.hist(bins=20, alpha=0.5)
```

#### Dla SalePrice także usuwamy outlinery, dając maksymalną kwotę na 500000.

```python
chosen_data = chosen_data[chosen_data.SalePrice < 500000]
```

#### Logarytmujemy, aby dane bardziej przypominały rozkład normalny.

```python
chosen_data['SalePrice'] = np.log1p(chosen_data["SalePrice"])
chosen_data['SalePrice'].plot.hist(bins=50, alpha=0.5)
```

#### Wizualizacja pomiędzy wybranymi argumentami, a ceną mieszkania. 

```python
fig, axs = plt.subplots(1, 3, sharey=True)
chosen_data.plot(kind='scatter', x='YearBuilt', y='SalePrice', ax=axs[0], figsize=(16, 8))
chosen_data.plot(kind='scatter', x='LotArea', y='SalePrice', ax=axs[1])
chosen_data.plot(kind='scatter', x='BedroomAbvGr', y='SalePrice', ax=axs[2])
```

### 4. Przetwarzanie danych

Nie normalizuje danych, gdyż po normalizacji i użyciu regresji liniowej otrzymałem błąd średniokwadratowy równy 0 dla każdego rodzaju regresji.

### 5. Trenowanie modeli

#### Podział danych na dane trenowane i testowane w celu wyeliminowania przeuczenia(overfitting).

```python
feature_cols = ['YearBuilt','LotArea','BedroomAbvGr']
sale_price = ["SalePrice"]

X_train = chosen_data[feature_cols][:-20]
y_train = chosen_data[sale_price][:-20]
X_test = chosen_data[feature_cols][:20]
y_test = chosen_data[sale_price][:20]
```

#### Zwykła regresja liniowa

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)

y_pred = regr.predict(X_test)

print('Współczynniki: \n', regr.coef_)
print("Średni błąd kwadratowy dla testowego zbioru: %.5f"
      % metrics.mean_squared_error(y_test,y_pred))

y_pred = regr.predict(X_train)
print("Średni kwadratowy błąd dla trenowanego zbioru: %.5f"
      % metrics.mean_squared_error(y_train,y_pred))
```

#### Regresja z regularyzacją L1 (Lasso)

```python
lasso_regr = linear_model.Lasso(alpha=1)
lasso_regr.fit(X_train,y_train)
y_pred = lasso_regr.predict(X_test)

print('Współczynniki: \n', lasso_regr.coef_)
print("Średni błąd kwadratowy dla testowego zbioru: %.5f"
      % metrics.mean_squared_error(y_test,y_pred))

y_pred = lasso_regr.predict(X_train)
print("Średni błąd kwadratowy dla trenowanego zbioru: %.5f"
      % metrics.mean_squared_error(y_train,y_pred))
```

#### Poszukiwanie najlepszej regresji Lasso od doboru atrybutu alpha

Wykres pokazuje, że im mniejsze alfa, tym lepiej dopasowana regresja.

```python
alphas = alphas = range(1,30) 
lassos = [metrics.mean_squared_error(linear_model.Lasso(alpha=alpha).fit(X_train,y_train).predict(X_test),y_test)
            for alpha in alphas]

lassos = Series(lassos, index = alphas)
lassos.plot(title = "Lasso")
plt.xlabel("alpha")
plt.ylabel("Średni błąd kwadratowy")

print("Najmniejszy średni błąd kwadratowy: %.5f"
      % metrics.mean_squared_error(linear_model.Lasso(alpha=0.0001).fit(X_train,y_train).predict(X_test),y_test))
```

#### Regresja z regularyzacją L2 (Ridge)

```python
ridge_regr = linear_model.Ridge(alpha=1)
ridge_regr.fit(X_train,y_train)
y_pred = ridge_regr.predict(X_test)

print('Współczynniki: \n', ridge_regr.coef_)
print("Średni błąd kwadratowy dla testowego zbioru: %.5f"
      % metrics.mean_squared_error(y_test,y_pred))

y_pred = ridge_regr.predict(X_train)

print("Średni błąd kwadratowy dla trenowanego zbioru: %.5f"
      % metrics.mean_squared_error(y_train,y_pred))
```

#### Poszukiwanie najlepszej regresji Lasso od doboru atrybutu alpha

Wykres pokazuje, że najmniejszy błąd jest mniej wiecej dla alfa równego 220.

```python
alphas = range(1,700) 
rigdes = [metrics.mean_squared_error(linear_model.Ridge(alpha=alpha).fit(X_train,y_train).predict(X_test),y_test)
            for alpha in alphas]

rigdes = Series(rigdes, index = alphas)
rigdes.plot(title = "Rigde")
plt.xlabel("alfa")
plt.ylabel("Średni błąd kwadratowy")

print("Najmniejszy średni błąd kwadratowy: %.5f"
      % metrics.mean_squared_error(linear_model.Ridge(alpha=220).fit(X_train,y_train).predict(X_test),y_test))

ridge_regr = linear_model.Ridge(alpha=220)
ridge_regr.fit(X_train,y_train)

print('Współczynniki: \n', ridge_regr.coef_)
```

#### Regresja Elastic Net

```python
net_regr = linear_model.ElasticNet(alpha=1,l1_ratio=0.5)
net_regr.fit(X_train,y_train)
y_pred = net_regr.predict(X_test)

print('Współczynniki: \n', net_regr.coef_)
print("Średni błąd kwadratowy dla testowego zbioru: %.5f"
      % metrics.mean_squared_error(y_test,y_pred))

y_pred = net_regr.predict(X_train)

print("Średni błąd kwadratowy dla trenowanego zbioru: %.5f"
      % metrics.mean_squared_error(y_train,y_pred))
```

#### Poszukiwanie najlepszej regresji Elastic Net od doboru atrybutu alpha

Wykres pokazuje, że im mniejsze alfa, tym lepiej dopasowana regresja.

```python
alphas = range(1,30) 
nets = [metrics.mean_squared_error(linear_model.ElasticNet(alpha=alpha).fit(X_train,y_train).predict(X_test),y_test)
            for alpha in alphas]

nets = Series(nets, index = alphas)
nets.plot(title = "Elastic Net")
plt.xlabel("alfa")
plt.ylabel("Średni błąd kwadratowy")

print("Najmniejszy średni błąd kwadratowy: %.5f"
      % metrics.mean_squared_error(linear_model.ElasticNet(alpha=0.0001,l1_ratio=0.5).fit(X_train,y_train).predict(X_test),y_test))

net_regr = linear_model.ElasticNet(alpha=0.0001,l1_ratio=0.5)
net_regr.fit(X_train,y_train)

print('Współczynniki: \n', net_regr.coef_)
```
